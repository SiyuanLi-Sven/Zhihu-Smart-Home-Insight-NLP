```python
#停用词与分词
def stopwordslist(filepath):
    stopwords = [line.strip() for line in open(filepath, 'r',encoding='utf-8').readlines()]
    return stopwords


stopwords = stopwordslist('stop_words.txt')
stopwords2 = stopwordslist('stopwordsChinese.txt')
```

```python
#新分词函数
def make_text_new(data):
    text =','.join([i for i in data['content']])
    words = jieba.cut(text)
    lines = []
    for word in words:
        if len(word) >= 2:
            lines.append(word)

    lines2 = []
    for x in lines:
        if (x not in stopwords) &(x not in stopwords2) &(x not in 
        ['疫情', '广州', '海珠', '花都区','海珠区', '核酸', '检测', '措施', '感染者', '防控', '无症状', '确诊', '新增', '希望','病例','10'
        ,'输入', '本土', '境外', '肺炎', '深圳', '新冠', '做好', '广州市', '全省', '采样', '13', '24', '12', '11', '报告', '2022'
        ,'冠状病毒', '新型', '情况', '区域', '街道', '社区', '视频', '输入', '肺炎', '新闻', '微博', '真的', '12', '11', '报告', '2022'
        ,'新华','48','居住']):
            lines2.append(x)

        
    # 6.获取词频最高的50个词
    result = Counter(lines2).most_common(50)
    #print(result)
    content = ' '.join(lines2)
    return lines2, result,  content
```

```python
#查看评论数在时间上的分布
data = pd.read_csv('washed_content_info_guangzhuo(6).csv')
data['datetime'] = '2022-'+data['month'].apply(lambda x: str(x))+'-'+data['day'].apply(lambda x: str(x))
data['time']=pd.to_datetime(data['datetime'])
```

```python
data0 = pd.DataFrame(data, columns=[ 'content','time'])
data0.dropna(axis=0,how="any",thresh= None,subset='content',inplace=True)
data0.info()
```

```python
#十月
listpic = []
for i in[0, 1, 2, 3 ,4 ,5]:
    topics = ['阳性','健康','口罩','居家','黄码','防疫','热线','放开','症状','感染','物资']
    a = np.timedelta64(5, 'D')
    
    if i <=4:
        dddd = data0[(data0['time']<np.datetime64('2022-10-01')+(i+1)*a)&(np.datetime64('2022-10-01')+i*a<=data0['time'])]
        lines, result, content = make_text_new(dddd)
        for j in topics:
            b_1 = '2022/10/'+str((5*i+1))
            b_2 = lines.count(j)
            b_3 = j
            b=[b_1, b_2, b_3]
            listpic.append(b)
    if i ==5:
        dddd = data0[(data0['time']<=np.datetime64('2022-10-31'))&(np.datetime64('2022-10-25')<=data0['time'])]
        lines, result, content = make_text_new(dddd)
        for j in topics:
            b_1 = '2022/10/25'
            b_2 = lines.count(j)
            b_3 = j
            b=[b_1, b_2, b_3]
            listpic.append(b)        



```

```python
def themeRiver():



    # x_data = ["小米10_青春版", "小米10_pro", "红米k30pro", "华为nova 7", "荣耀X10", "荣耀30S"]
    x_data= ['阳性','健康','口罩','居家','黄码','防疫','热线','放开','症状','感染','物资']
    y_data = listpic

    tr= (
        ThemeRiver(init_opts=opts.InitOpts(width="1300px", height="700px"))
            .add(
            series_name=x_data,
            data=y_data,
            singleaxis_opts=opts.SingleAxisOpts(
                pos_top="50", pos_bottom="50", type_="time"
            ),
        )
            .set_global_opts(
            tooltip_opts=opts.TooltipOpts(trigger="axis", axis_pointer_type="line"),
            legend_opts=opts.LegendOpts(pos_left='50%')
        )

    )
    return tr
```
